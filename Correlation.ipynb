# Проверим корреляцию между IT компаниями

```python
import yfinance as yf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Список тикеров для исследования из различных отраслей
tickers = [ 'MSFT', 'NVDA', 'ADBE', 'INTC', 'CSCO']

# Получение исторических данных цен акций
data = yf.download(tickers, start='2010-01-01', end='2022-12-31')['Adj Close']

# Вычисление корреляции
correlation_matrix = data.corr()

# Визуализация корреляций
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Stocks')
plt.show()
```
